```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
```

```python
pulsar_data = pd.read_csv("pulsar_stars.csv")
orig_column_names = list(pulsar_data)
new_column_names = []
for i in orig_column_names:
    new_column_names.append('_'.join(i.strip().split(' ')))
pulsar_data.columns = new_column_names
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(pulsar_data, pulsar_data["target_class"]):
    strat_train_set = pulsar_data.loc[train_index]
    strat_test_set = pulsar_data.loc[test_index]
```

```python
pulsar_train_X = strat_train_set.drop('target_class', axis=1)
pulsar_train_y = strat_train_set['target_class']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_train_X = scaler.fit_transform(pulsar_train_X)
```

```python
sgd_clf = SGDClassifier(random_state=42)
```

```python
sgd_clf.fit(scaled_train_X, pulsar_train_y)
```
